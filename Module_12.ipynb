# Credit Risk Classification

```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df =  pd.read_csv('../Desktop/csv/lending_data.csv')

# Review the DataFrame
df
```

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = df['loan_status']

# Separate the X variable, the features
X = df.drop(columns='loan_status')
```

```python
# Review the y variable Series
y.head()
```

```python
# Review the X variable DataFrame
X.head()
```

```python
# Check the balance of our target values
df.value_counts('loan_status')
```

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

## Create a Logistic Regression Model with the Original Data

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
lr_df = LogisticRegression(random_state = 1)

# Fit the model using training data
lr_df.fit(X_train, y_train)
```

```python
# Make a prediction using the testing data
y_pred = lr_df.predict(X_test)
```

```python
# Print the balanced_accuracy score of the model
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Print the classification report for the model
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** Through the data given via the classification report it is evident to suggest that both 'healthy loan' and 'high risk-loan' are well predicted by the logisitic regression model, with accuracy values of 100% for healthy loans and 88% for high risk loans.
